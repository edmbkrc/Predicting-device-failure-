```python
df_dummies=pd.get_dummies(df,drop_first=1)
```

```python
df_dummies
```

```python
y=df["failure"]
```

```python
y
```

```python
x=df_dummies.drop(["failure"],axis=1)
```

```python
x
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
x=scaler.fit_transform(x)
```

```python
x
```

```python
#pip install --use-deprecated=legacy-resolver pycaret[full]
```

```python
from imblearn.under_sampling import RandomUnderSampler#imbalanced verilerde kullanılır
from sklearn.ensemble import GradientBoostingClassifier
gbc=GradientBoostingClassifier()
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 42)
```

```python

rus = RandomUnderSampler(sampling_strategy = "majority")
x_rus, y_rus = rus.fit_resample(x_train,y_train)
```

```python
gbc.fit(x_rus,y_rus)
```

```python
pred=gbc.predict(x_test)
```

```python
confusion_matrix(y_test,pred)
```

```python
accuracy_score(y_test,pred),precision_score(y_test,pred),recall_score(y_test,pred),f1_score(y_test,pred)
```

```python
##DEEP LEARNING
```

```python
import pandas as pd 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_rus.shape
```

```python
y_rus.shape
```

```python
model=Sequential()
model.add(Dense(512,activation="relu"))
model.add(Dense(512,activation="relu"))
model.add(Dense(512,activation="relu"))
model.add(Dense(512,activation="relu"))
model.add(Dense(512,activation="relu"))
model.add(Dense(512,activation="relu"))
model.add(Dense(1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
model.fit(x_rus,y_rus,epochs=300,batch_size=10,verbose=1 )
model.summary()
```

```python
scores=model.evaluate(x_rus,y)
print('%s:%.2f%%' % (model.metrics_names[1],scores[1]*100))
```
